```python
Classes = {
    0.0 : 'N',
    1.0 : 'S',
    2.0 : 'V',
    3.0 : 'F',
    4.0 : 'Q'
}

df_test['label'] = df_test['label'].map(Classes)
df_test
```

```python
X_train = df_train.drop('label', axis=1)
X_train
```

```python
y_train = df_train.label
y_train
```

```python
X_test = df_test.drop('label', axis=1)
X_test
```

```python
y_test = df_test.label
y_test
```

```python
Y_train = pd.get_dummies(y_train, columns = ['N', 'S', 'V', 'F', 'Q'], drop_first=True)
Y_train
```

```python
Y_test = pd.get_dummies(y_test, columns = ['N', 'S', 'V', 'F', 'Q'], drop_first=True)
Y_test
```

```python
Y_train.shape[1]
```

```python
Y_test.shape[1]
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Dense, Input, Flatten
```

```python
input_shape = (X_train.shape[1], 1,)
input_shape
```

```python
model = Sequential([
    Input(shape=input_shape),
    Conv1D(64, (5), activation='relu'),
    MaxPooling1D(pool_size=5),
    
    Conv1D(64, (5), activation='relu'), 
    MaxPooling1D(pool_size=5),
    
    Flatten(),
    Dense(32, activation='relu'),
    Dense(4, activation='softmax')
])
```

```python
opt = keras.optimizers.Adam(learning_rate=0.00003)
model.compile(
    optimizer=opt,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
model.fit(X_train, Y_train, epochs=10)
```

```python
model.evaluate(X_test, Y_test)
```

```python
model.save("..//models//01_Perfect_alpha.h5")
```
